```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/Real estate.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,2:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_scaled=mms.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=5)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
knr=KNeighborsRegressor(n_neighbors=5)
dtr=DecisionTreeRegressor()
abr=AdaBoostRegressor()
rfr=RandomForestRegressor(random_state=1,max_depth=8)
xgb=XGBRegressor()
models=[knr,dtr,abr,rfr,xgb]
for i in models:
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(r2_score(y_test,y_pred)*100)
```

```python
#highest in RandomForestRegressor call seperately
rfr.fit(X_train,y_train)
y_pred_rfr=rfr.predict(X_test)
y_pred_rfr
```

```python
print(r2_score(y_test,y_pred_rfr)*100)
```
